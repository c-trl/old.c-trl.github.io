#Cleaning Data

###Replace all NaN in columns '1' through '5' with 0

```python
df[['1','2','3','4','5']] = df[['1','2','3','4','5']].fillna(0)
```

```python
df.describe()
```

#Comparing interest against other factors: Industry

```python
ind = df[['Interest', 'Industry']]
```

```python
#sum of interest in job leads
indInterest = ind.groupby('Industry').sum()
indInterest = indInterest.reset_index()
indInterest
```

```python
#total leads found by industry
indLeads = ind.groupby('Industry').count()
indLeads = indLeads.reset_index()
indLeads
```

```python
industry_interest = indLeads.merge(indInterest, on="Industry")

#renaming columns
industry_interest.columns = ['Industry', 'Leads Found', 'Total Interest']

#creating a column with averages
industry_interest['Average Interest per Lead'] = industry_interest['Total Interest']/industry_interest['Leads Found']

#plotting sorted data
industry_interest[['Industry', 'Leads Found', 'Total Interest']].sort(
columns='Total Interest', ascending=False).plot(
kind='bar', figsize=(30,10), x='Industry', title="The Job Hunt")
```

```python
industry_interest[['Industry', 'Average Interest per Lead']].sort(
columns='Average Interest per Lead', ascending=False).plot(
kind='bar', figsize=(30,10), x='Industry', title="Average Interest per Job Lead", xlim=(-0.5,7.5), color=colors, legend=False)
```

#Comparison of Interests in Job Leads by Industry

###Job Lead Interest: Very High

```python
five = df[['5', 'Industry']]
five = five[five['5'] == 1].groupby('Industry').count()
five = five.reset_index()
five = five.sort(columns='5', ascending=False)
five
```

```python
five.plot(kind='bar', x='Industry', figsize=(30,10), xlim=(-0.5, 7.5), legend=False, title='Job Lead Interest: Very High', color=colors)
```

###Job Lead Interest: High

```python
four = df[['4', 'Industry']] #create a new df from the master df
four['4'] = four['4']-df['5'] #if a job lead was given a 5, subtract it from the list of job leads given a 4
four = four[four['4'] == 1].groupby('Industry').count() #group only leads given a 4, by industry
four = four.reset_index() #push index into df
four = four.sort(columns='4', ascending=False) #sort df by interest
four 
```

```python
four.plot(kind='bar', 
          x='Industry', 
          figsize=(30,10), 
          xlim=(-0.5, 7.5), 
          ylim=(0,35), 
          legend=False, 
          title='Job Lead Interest: High', 
          color=colors)
```

 ###Job Lead Interest: High-Very High

```python
min_four = df[['4', 'Industry']]
min_four = min_four[min_four['4'] == 1].groupby('Industry').count()
min_four = min_four.reset_index()
min_four = min_four.sort(columns='4', ascending=False)
min_four
```

```python
min_four.plot(kind='bar', 
          x='Industry', 
          figsize=(30,10), 
          xlim=(-0.5, 7.5), 
          ylim=(0,65), 
          legend=False, 
          title='Job Lead Interest: High-Very High', 
          color=colors)
```

#Share of Job Leads by Interest Level, All Industries

```python
byint = df[['1','2','3','4','5']]

#capping each interest level so that each interest level will only count leads given each resepctive interest level and not inclusive of higher levels
byint['4'] = byint['4']-df['5'] 
byint['3'] = byint['3']-df['4'] 
byint['2'] = byint['2']-df['3'] 
byint['1'] = byint['1']-df['2'] 

#renaming columns
byint.columns = ['Very Low', 'Low', 'Neutral', 'High', 'Very High']
byint.sum()
```

```python
byint.sum().plot(kind='pie', figsize=(15,15), autopct='%1.2f%%', explode = (0, 0, 0.05, 0.1, 0.15), colors=colors, shadow=True, startangle=90, title='Share of Job Leads by Interest in Lead')
```
